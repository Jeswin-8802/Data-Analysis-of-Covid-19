```python
X, Y, Z = {}, {}, {}
for i in DF.iloc[:,7:8].iterrows():
  if i[0][1] != '-': X.update({i[0][0]+', '+i[0][1] : i[1][0]})
  else: X.update({i[0][0]:  i[1][0]})
for i in DF.iloc[:,15:16].iterrows():
  if i[0][1] != '-': Y.update({i[0][0]+', '+i[0][1] : i[1][0]})
  else: Y.update({i[0][0]:  i[1][0]})
for i in DF.iloc[:,23:24].iterrows():
  if i[0][1] != '-': Z.update({i[0][0]+', '+i[0][1] : i[1][0]})
  else: Z.update({i[0][0]:  i[1][0]})

A, B, C = X.keys(), Y.keys(), Z.keys()

# storing the values of max (Confirmed, Recovered, Deaths) for each of the country/region
```

```python
from operator import itemgetter
X = dict(sorted(X.items(), key=itemgetter(1), reverse = True))
Y = dict(sorted(Y.items(), key=itemgetter(1), reverse = True))
Z = dict(sorted(Z.items(), key=itemgetter(1), reverse = True))

print(X, Y, Z, sep = '\n----------------------------------------\n')

# Here we are sorting Country/Region and Province/State according to the three criterias (Confirmed, Recovered, Deaths) in descending order to see which Country/Region has the highest of all of the three measures
```

```python
l1, l2, l3 = [], [], []
for i in X: l1.append(i)
for i in Y: l2.append(i)
for i in Z: l3.append(i)

# we are getting only the names of countries in order wrt the three criterias
```

# ***`Now we perform Regression Analysis and predict its accuracy`*** 

Performing Simple Linear Regression taking World Column as predictor and Mortality as dependent

```python
# we now add a new column to the dataframe created above which stores the number of deaths per day in the entire world

temp = df.groupby(['Country/Region','Province/State'])['Deaths'].apply(np.array)
l = np.array([abs(sum([temp[j][i+1] for j in range(len(temp))]) - sum([temp[j][i] for j in range(len(temp))])) for i in range(len(temp[0]) - 1)])
new_df['Mortality'] = l
DF = new_df[['World', 'Mortality']]
DF
```

```python
# Visualisation

plt.figure(figsize=(18, 10))

plt.xlabel('Confirmed Cases Worldwide')
plt.ylabel('Worldwide Mortality rate')
plt.scatter(DF['World'].values, DF['Mortality'].values, color = 'black')

plt.show()
```

```python
plt.figure(figsize=(15, 7))
plt.subplot(1, 3, 1)
plt.boxplot(DF['World'].values, labels = ['Confirmed Cases Worldwide'])

plt.subplot(1, 3, 2)
plt.boxplot(DF['Mortality'].values, labels = ['Worldwide Mortality Rate'])
plt.show()
```

```python
DF = DF.drop(DF[DF['Mortality']>20000].index) # threshhold -> 20,000
DF = DF.drop(DF[DF['World']>1000000].index) # threshhold ->1,000,000
new_df = new_df.drop(new_df[new_df['World']>1000000].index) # for later

# removing outliers obtained through the plot above
plt.figure(figsize=(15, 7))
plt.subplot(1, 3, 1)
plt.boxplot(DF['World'].values, labels = ['Confirmed Cases Worldwide'])

plt.subplot(1, 3, 2)
plt.boxplot(DF['Mortality'].values, labels = ['Worldwide Mortality Rate'])
plt.show()
```

```python
x = DF.iloc[:,[0]].values
y = DF['Mortality'].values

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x, y)
```

```python
y_pred = model.predict(x)

# visualization of legression line

plt.figure(figsize=(18, 10))

plt.xlabel('Confirmed Cases Worldwide')
plt.ylabel('Worldwide Mortality rate')
plt.plot(x, y_pred, color = 'orange', linewidth = 4)
plt.scatter(DF['World'].values, DF['Mortality'].values, color = 'grey')

plt.show()
```

```python
model.coef_[0], model.intercept_

# y = model.coef_ * x + model.intercept_
```

```python
100 - sum(abs(y - y_pred)) * 100 / sum(y)

# accuracy
```

```python
del new_df['World']
# we are deleting the column world as we will not use it to predict the number of deaths
```

```python
new_df
```

```python
x = new_df.iloc[:,0:280].values
y = new_df.iloc[:,280].values
```

```python
from sklearn.model_selection import train_test_split
import numpy as np
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 0)

print(x_train.shape, x_test.shape, y_train.shape, y_test.shape, sep = '\n')
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
df_t = pd.DataFrame({'Predicted' : y_pred,
                     'Known' : y_test})
df_t = df_t.astype({'Predicted':int})
df_t
```

```python
df_t.describe()
```

```python
# we find that there is an outliers such as extremely high values and -ve values therefore we calculate the accuracy after removing it
for i in range(len(df_t)):
  if df_t.loc[i][0]<0 or df_t.loc[i][0]>100000:
    df_t = df_t.drop(i)
    i -= 1
```

```python
plt.figure(figsize=(25, 10))
plt.ylabel('Mortality')
plt.bar(np.array(df_t.index), df_t['Predicted'], label = 'Predicted', color = 'grey')
plt.bar(np.array(df_t.index), df_t['Known'], label = 'Known', color = 'black')
plt.xticks(df_t.index[::10])
plt.legend()
plt.show()
```

```python
import math
100 - (math.sqrt(sum((df_t['Predicted'] - df_t['Known'])**2)) * 100 / sum(y_test))

# accuracy (RMS)
```
